````python
import csv
import ast
import math
N_ch = 8
with open("data/recipes_full.csv", 'r') as file:
    csvreader = csv.reader(file)
    header = next(csvreader)
    lines = []
    for row in csvreader:
        row_id = row[1]
        row_steps = row[6]
        row_tags = ast.literal_eval(row[5]) # list of tags
        for tag in row_tags:
            lines.append([row_id, tag, row_steps])
    n_total = len(lines)
    
    def write_chunk(part, chunk):
        with open('data/chunks/id_tag_nsteps_'+ str(part) +'.csv', 'w') as chunk_file:
            writer = csv.writer(chunk_file)
            writer.writerow(['id', 'tag', 'n_steps'])
            writer.writerows(chunk)
            
            
    n_chunk = int(len(lines) / N_ch)
    for i in range(0, N_ch):
        if i == N_ch - 1:
            write_chunk(i + 1, lines[i * n_chunk:])
        else:
            write_chunk(i + 1, lines[i * n_chunk: i * n_chunk + n_chunk])
````

2. Напишите функцию, которая принимает на вход название файла, созданного в результате решения задачи 1, считает среднее значение количества шагов для каждого тэга и возвращает результат в виде словаря.

````python
#using pandas

def mean_steps_pandas(path):
    df = pd.read_csv(path)
    return df.groupby('tag').mean()['n_steps'].round(3).to_dict()
````

````python
file_name = "id_tag_nsteps_1.csv"
res_1 = mean_steps_pandas('data/chunks/' + file_name)
````

````python
# using csvreader - 1 defaultdict
from collections import defaultdict

def mean_steps(path):
    tag_sum_num = defaultdict(lambda:[0, 0]) #first: sum, secord: number
    
    with open(path, 'r') as file:
        dictfile = csv.DictReader(file)
        for row in dictfile:
            tag_sum_num[row['tag']][0] += int(row['n_steps'])
            tag_sum_num[row['tag']][1] += 1

    res_dict = defaultdict(float)
    for key, values in tag_sum_num.items():
        res_dict[key] = round(values[0] / values[1], 3)
    return dict(res_dict)


res_2 = mean_steps("data/chunks/id_tag_nsteps_1.csv")
````

````python
# using csvreader - using 2 deafualtdicts
from collections import defaultdict

def mean_steps_2(path):
    tag_sum = defaultdict(int)
    tag_num = defaultdict(int)
    with open(path, 'r') as file:
        dictfile = csv.DictReader(file)
        for row in dictfile:
            tag_sum[row['tag']] += int(row['n_steps'])
            tag_num[row['tag']] += 1

    res_dict = defaultdict(float)
    for key in tag_sum:
        res_dict[key] = round(tag_sum[key] / tag_num[key], 3)
    return dict(res_dict)


res_3 = mean_steps_2("data/chunks/id_tag_nsteps_1.csv")
````

3. Напишите функцию, которая считает среднее значение количества шагов для каждого тэга по всем файлам, полученным в задаче 1, и возвращает результат в виде словаря. Не используйте параллельных вычислений. При реализации выделите функцию, которая объединяет результаты обработки отдельных файлов. Модифицируйте код из задачи 2 таким образом, чтобы иметь возможность получить результат, имея результаты обработки отдельных файлов. Определите, за какое время задача решается для всех файлов.


````python
def sum_num(path):
    tag_sum = defaultdict(int)
    tag_num = defaultdict(int)
    with open(path, 'r') as file:
        dictfile = csv.DictReader(file)
        for row in dictfile:
            tag_sum[row['tag']] += int(row['n_steps'])
            tag_num[row['tag']] += 1
    return (dict(tag_sum), dict(tag_num))
````

````python
from collections import Counter
def total_mean_steps(number):
    res_list = [sum_num(f'data/chunks/id_tag_nsteps_{i+1}.csv') for i in range(0, number)]
        
    total_sum = defaultdict(int)
    total_num = defaultdict(int)
    for buf_sum, buf_num in res_list:
        total_sum = dict(Counter(total_sum) + Counter(buf_sum))
        total_num = dict(Counter(total_num) + Counter(buf_num))
    
    res_dict = defaultdict(float)
    for key in total_sum:
        res_dict[key] = round(total_sum[key] / total_num[key], 3)
    return dict(res_dict)
````

````python
total_mean_steps(8)
````

````python
import time
start_time = time.time()
total_mean_steps(8)
print(f'{time.time() - start_time} seconds')
````

4. Решите задачу 3, распараллелив вычисления с помощью модуля `multiprocessing`. Для обработки каждого файла создайте свой собственный процесс. Определите, за какое время задача решается для всех файлов.

````python
import multiprocessing as mp

def sum_num_mult(path, output):
    tag_sum = defaultdict(int)
    tag_num = defaultdict(int)
    with open(path, 'r') as file:
        dictfile = csv.DictReader(file)
        for row in dictfile:
            tag_sum[row['tag']] += int(row['n_steps'])
            tag_num[row['tag']] += 1
    output.put((dict(tag_sum), dict(tag_num)))

def total_mean_steps_mult(number):
    output = mp.Queue()
    processes = [mp.Process(target=sum_num_mult, args=(f'data/chunks/id_tag_nsteps_{i+1}.csv', output))
                     for i in range(0, number)]
    for p in processes:
        p.start()
    
    for p in processes:
        p.join()
    
    res_list = [output.get() for p in processes]
    print('here')
    total_sum = defaultdict(int)
    total_num = defaultdict(int)
    for buf_sum, buf_num in res_list:
        total_sum = dict(Counter(total_sum) + Counter(buf_sum))
        total_num = dict(Counter(total_num) + Counter(buf_num))
    
    res_dict = defaultdict(float)
    for key in total_sum:
        res_dict[key] = round(total_sum[key] / total_num[key], 3)
    return dict(res_dict)
````

````python
import time
start_time = time.time()
total_mean_steps_mult(8)
print(f'{time.time() - start_time} seconds')
````
